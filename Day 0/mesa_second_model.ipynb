```python
from model import ForestFire
```

```python
# Create a 50x50 forest with 60% tree density
model = ForestFire(width=50, height=50, density=0.6)

# Run for 20 steps
for i in range(20):
    model.step()
    
    # We can use AgentSet's built-in selection filters now too!
    burning_trees = len(model.agents.select(lambda t: t.condition == "On Fire"))
    print(f"Step {i}: {burning_trees} trees on fire")
```

```python
from mesa.visualization import SolaraViz, make_space_component

def agent_portrayal(agent):
    """
    Maps an agent's state to a visual representation (dictionary).
    's' stands for square marker.
    """
    if agent.condition == "Fine":
        return {"color": "green", "marker": "s", "size": 50}
    elif agent.condition == "On Fire":
        return {"color": "red", "marker": "s", "size": 50}
    elif agent.condition == "Burned Out":
        return {"color": "gray", "marker": "s", "size": 50}

# Instantiate the model
model = ForestFire(width=50, height=50, density=0.6)

# Create the Solara component
page = SolaraViz(
    model,
    components=[make_space_component(agent_portrayal)],
    name="Forest Fire Simulation"
)

# Display the UI (Jupyter will automatically render this interactive widget)
page
```
